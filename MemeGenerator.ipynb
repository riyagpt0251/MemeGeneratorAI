```python
# Step 3: Download Meme Templates (or upload your own)
# You can upload your own meme templates to Colab or use these sample URLs
!wget https://i.imgflip.com/1bij.jpg -O distracted_boyfriend.jpg
!wget https://i.imgflip.com/30b1gx.jpg -O drake_hotline_bling.jpg
!wget https://i.imgflip.com/1g8my4.jpg -O two_buttons.jpg

# Step 4: Define Meme Templates
MEME_TEMPLATES = {
    "distracted_boyfriend": "distracted_boyfriend.jpg",
    "drake_hotline_bling": "drake_hotline_bling.jpg",
    "two_buttons": "two_buttons.jpg",
}

# Step 5: Function to Add Text to Meme
def create_meme(template_name, top_text, bottom_text=""):
    """
    Create a meme by adding text to a template.

    Parameters:
    - template_name (str): Name of the meme template.
    - top_text (str): Text to add at the top of the meme.
    - bottom_text (str): Text to add at the bottom of the meme.

    Returns:
    - PIL Image object.
    """
    # Load the template image
    if template_name not in MEME_TEMPLATES:
        raise ValueError(f"Template '{template_name}' not found. Available templates: {list(MEME_TEMPLATES.keys())}")

    image_path = MEME_TEMPLATES[template_name]
    image = Image.open(image_path)

    # Initialize ImageDraw
    draw = ImageDraw.Draw(image)

    # Load a font (you can also upload your own font file)
    try:
        font = ImageFont.truetype("arial.ttf", 40)
    except IOError:
        font = ImageFont.load_default()

    # Add top text
    text_position_top = (image.width // 2, 10)  # Centered at the top
    draw.text(text_position_top, top_text, font=font, fill="white", anchor="mt", stroke_width=2, stroke_fill="black")

    # Add bottom text
    text_position_bottom = (image.width // 2, image.height - 50)  # Centered at the bottom
    draw.text(text_position_bottom, bottom_text, font=font, fill="white", anchor="mb", stroke_width=2, stroke_fill="black")

    return image

```

```python
# Step 6: Generate and Display a Meme
template_name = "distracted_boyfriend"  # Choose a template
top_text = "Me: Working on my project"
bottom_text = "Also me: Procrastinating on memes"

meme = create_meme(template_name, top_text, bottom_text)

# Display the meme
display(meme)

# Save the meme
meme.save("generated_meme.png")
print("Meme saved as 'generated_meme.png'")
```
